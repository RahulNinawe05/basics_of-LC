```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_text_splitters import CharacterTextSplitter

loader = TextLoader("speech.txt")
document = loader.load()
text_splitter=CharacterTextSplitter(chunk_size=1000, chunk_overlap=50)
docs = text_splitter.split_documents(document)
```

```python
docs
```

```python
embedding=HuggingFaceEmbeddings()
db= FAISS.from_documents(docs, embedding)
db
```

```python
## querying
query = "Which principles does the speaker claim America is fighting to defend?"
doc= db.similarity_search(query)
doc[0].page_content
```

## AS a Retriever
we can also convert the vectorsotore into a retrivar class. this allows us to easily use it in other langchain methods which laegely work eith retrievers

```python
# Retriever - kisi data aise use karna jahan se LLM sirf relevant(context) information retrieve (nikal) sake, pura data nahi.
retriver=db.as_retriever()
docse = retriver.invoke(query)
docse[0].page_content
```

## Similarity Search with Score
Similarity Search with Score is the process of finding the most relevant items (like text chunks, images, or embeddings) to a given query and returning them along with a numeric score that shows how similar they are.

```python
docs_with_similar_score = db.similarity_search_with_score(query)
docs_with_similar_score
```

```python
embedding_vector = embedding.embed_query(query)
embedding_vector
```

```python
docs_score = db.similarity_search_by_vector(embedding_vector)
docs_score
```

```python
### saving and loading 
db.save_local("faiss_index")
```

```python
new_df= FAISS.load_local("faiss_index", embedding, allow_dangerous_deserialization=True)
docs = new_df.similarity_search(query)
docs
```
