## Custom Tool

#### Method 1 => using @tool decorator

```python
from langchain_core.tools import tool
# step 1 - Create a Function

def multiply(a,b):
    """multiply two number"""
    return a * b
```

```python
# step 2 - type hinting

def multiply(a: int,b: int)-> int:
    """multiply two number"""
    return a * b
```

```python
# step 3 - add @tool decorator

@tool
def multiply(a: int, b:int)-> int:
    """multiply two number"""
    return a * b 
```

```python
result = multiply.invoke({'a':3, 'b':4})
print(result)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
print(multiply.args_schema.model_json_schema())
```

#### Method 2 => using StructureTool & Pydantic
- A input to the follow a Structure Schema using Pydantic model

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel,Field
```

```python
class multiply_type(BaseModel):
    a: int = Field(description="the First Number")
    b: int = Field(description="The Second Number")
```

```python
def multiply_tool(a, b):
    return a * b
```

```python
Mul_tool = StructuredTool.from_function(
    func=multiply_tool,
    name='MULTILICATION',
    description= "The multiplication of two tools",
    args_schema = multiply_type
)
```

```python
data = Mul_tool.invoke({'a':2, 'b':4})
print(data)
```

```python
print(Mul_tool.name)
print(Mul_tool.description)
print(Mul_tool.args)
```

#### Method 3 => Using BaseTool Class
- LangChain ka main base class jisse sab tools bante hain.
Yeh define karta hai ki ek tool input lega, process karega aur output return karega.

```python
from langchain_core.tools import BaseTool
from typing import Type
```

```python
class multipe(BaseModel):
    a: int = Field(description="the First Number")
    b: int = Field(description="The Second Number")
```

```python
class Multiple_tool(BaseTool):
    name: str = "MULTIPLE"
    description: str = "mulitply two number"

    args_schema : type[BaseModel] = multipe

    def _run(self, a:int, b:int) -> int:
        return a * b
    
```

```python
re_tool = Multiple_tool()
```

```python
data = re_tool.invoke({'a':2, 'b':4})
print(data)
```
