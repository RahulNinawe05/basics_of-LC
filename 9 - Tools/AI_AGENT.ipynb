```python
from dotenv import load_dotenv
import os
load_dotenv()
from langchain_groq import ChatGroq
```

```python
llm = ChatGroq(
    api_key=os.getenv("GROQ_API_KEY"),
    model_name="llama-3.1-8b-instant"
)
```

```python
# Step - 1

# tool-1
from langchain_community.tools import DuckDuckGoSearchRun

search_tool = DuckDuckGoSearchRun()

# result = search_tool.invoke("womens cricket")
# print(result)


#tool-2
from langchain.tools import tool
import requests
@tool

def weather_tool(city: str) -> str:
    """this function fetch the currunt weather data"""

    API_KEY = os.getenv("WEATHER_API_KEY")

    url = f"https://api.weatherapi.com/v1/current.json?key={API_KEY}&q={city}&aqi=yes"

    response = requests.get(url=url)

    return response.json()

@tool
def stock_price_tool(stock: str) -> str:
    """The fuction fetch the data currnet Stock Prce """

    url = f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={stock}&interval=5min&apikey=demo"

    response = requests.get(url=url)

    return response.json()
```

```python
from langchain.agents import create_react_agent, AgentExecutor
from langchain import hub
```

```python
# Step -2
prompt = hub.pull("hwchase17/react")
```

```python
# Step -3
AI_Agent = create_react_agent(
    llm=llm,
    tools=[search_tool,weather_tool,stock_price_tool],
    prompt=prompt
)
```

```python
# Step -4
Agent_Executor = AgentExecutor(
    agent=AI_Agent,
    tools=[search_tool,weather_tool,stock_price_tool],
    verbose=True # show the Thought Trace
)
```

```python
response = Agent_Executor.invoke({'input':"who owner of zoho "})
print(response)
```
