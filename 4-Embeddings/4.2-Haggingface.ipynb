## Sentence Transformer on Hagging Face
- Sentence Transformers are models that convert sentences, paragraphs, or documents into dense vector embeddings.
- These embeddings capture semantic meaning, so similar sentences have similar vectors.
- They are built on top of BERT / RoBERTa / DistilBERT and optimized for semantic search, clustering, and similarity tasks.

```python
# creat a Embeddign by huggingface
from langchain_huggingface import HuggingFaceEmbeddings
embedding = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text="Hey My Name is Rahul, I am learn deep drive in ML&DL"
qurey_result=embedding.embed_query(text)
qurey_result
```
